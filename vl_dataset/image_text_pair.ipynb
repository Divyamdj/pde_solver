```python
import h5py
import random

def create_pairs_from_files(files, descriptions, field="density"):
    pairs = []

    for path in files:
        try:
            with h5py.File(path, "r") as f:
                for t in range(101):
                    
                    field = pde_to_tensor(f["t0_fields"]["density"], traj_idx=0, time_idx=t)
                    text = random.choice(descriptions)

                    pairs.append({
                        "tensor": field,
                        "text": text,
                        "source_file": os.path.basename(path)
                    })
        except Exception as e:
            print("Skipping corrupted file:", path)

    return pairs
```

```python
train_pairs = create_pairs_from_files(train_files, descriptions)
val_pairs   = create_pairs_from_files(val_files, descriptions)
test_pairs  = create_pairs_from_files(test_files, descriptions)

print("Train pairs:", len(train_pairs))
print("Val pairs:", len(val_pairs))
print("Test pairs:", len(test_pairs))
```

```python
# training data

import os
import json

os.makedirs("tensors", exist_ok=True)

with open("annotations.jsonl", "w") as f:
    for i, p in enumerate(train_pairs):
        np.save(f"tensors/{i}.npy", p["tensor"])
        record = {
            "tensor": f"tensors/{i}.npy",
            "text": p["text"]
        }

        f.write(json.dumps(record) + "\n")
```

```python
# testing data

import os
import json

os.makedirs("tensors_test", exist_ok=True)

with open("annotations_test.jsonl", "w") as f:
    for i, p in enumerate(test_pairs):
        np.save(f"tensors_test/{i}.npy", p["tensor"])
        record = {
            "tensor": f"tensors_test/{i}.npy",
            "text": p["text"]
        }

        f.write(json.dumps(record) + "\n")
```
