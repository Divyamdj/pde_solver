```python
import numpy as np
from PIL import Image
```

```python
def pde_to_rgb(data4d, t=0):
    arr = np.array(data4d[t, 0:3], dtype=np.float32)  # (3, 128, 384)
    arr = np.transpose(arr, (1, 2, 0))  # -> (128, 384, 3)

    arr = (arr - arr.min()) / (arr.max() - arr.min() + 1e-8)
    arr = (arr * 255).astype(np.uint8)

    return Image.fromarray(arr, mode="RGB")
```
